```python
import pandas as pd

# Load the uploaded CSV file 
data = pd.read_csv(r"C:\Users\DELL\Desktop\Python Related\survey.csv")

# Display basic information about the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Plot age distribution
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot gender distribution
plt.figure(figsize=(8, 5))
gender_counts = data['Gender'].value_counts()
gender_counts.plot(kind='bar', color=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot treatment response distribution
plt.figure(figsize=(8, 5))
treatment_counts = data['treatment'].value_counts()
treatment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Treatment Response Distribution')
plt.ylabel('')
plt.show()
```

```python
# Analyze relation between work interference and treatment
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='work_interfere', hue='treatment', palette='Set2')
plt.title('Work Interference vs. Treatment')
plt.xlabel('Work Interference')
plt.ylabel('Count')
plt.show()
```
